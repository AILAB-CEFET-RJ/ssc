```python
fig, ax0 = plt.subplots(figsize=FIGSIZE)
ax0 = sns.barplot(data=dfall, y='Model', x='HomogeneityScore', hue='Dataset', ax=ax0, seed=0)
ax0.set_xlim([0, 1.1])
ax0.set_ylabel('')
fig.tight_layout()
# plt.savefig(f'{FIGURES_FOLDER}/test/homogeneity-score.{EXTENSION}')
```

```python
def print_intervals(ax, decimals = 2):
        i = 1
        for i in range(len(ax.lines)//3):
                # small dataset
                errorbar = ax.lines[i]
                bar = ax.patches[i]
                l,u = errorbar.get_xdata()
                x = bar.get_width()
                p = (i+1) % 12 if (i+1) % 12 != 0 else 12
                print(f"p{p}: {x:.{decimals}f} [{l:.{decimals}f}, {u:.{decimals}f}]")

                # medium dataset
                i += 12
                errorbar = ax.lines[i]
                bar = ax.patches[i]
                l,u = errorbar.get_xdata()
                x = bar.get_width()
                p = (i+1) % 12 if (i+1) % 12 != 0 else 12
                print(f"p{p}: {x:.{decimals}f} [{l:.{decimals}f}, {u:.{decimals}f}]")

                # large dataset
                i += 12
                errorbar = ax.lines[i]
                bar = ax.patches[i]
                l,u = errorbar.get_xdata()
                x = bar.get_width()
                x = round(bar.get_width(), decimals)
                p = (i+1) % 12 if (i+1) % 12 != 0 else 12
                print(f"p{p}: {x:.{decimals}f} [{l:.{decimals}f}, {u:.{decimals}f}]\n")
```

## Wall Time

```python
fig, axs = plt.subplots(2, 1, figsize=FIGSIZE)

ax0 = sns.barplot(data=dfall[dfall['Dataset'] == 'small'], y='Model', x='WallTimeInSeconds', hue='Dataset', ax=axs[0], seed=0)
ax0.set_xlabel('seconds')
ax0.set_ylabel('')

handles0, labels0 = ax0.get_legend_handles_labels()
ax0.get_legend().remove()


colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

ax1 = sns.barplot(data=dfall[dfall['Dataset'] != 'small'], y='Model', x='WallTimeInHours', hue='Dataset', ax=axs[1], palette=colors[1:3], seed=0)
ax1.set_xlabel('hours')
ax1.set_ylabel('')

handles1, labels1 = ax1.get_legend_handles_labels()
ax1.get_legend().remove()

ax1.legend(handles0+handles1, labels0+labels1, loc='upper right')

fig.tight_layout()
# plt.savefig(f'{FIGURES_FOLDER}/test/wall-time.{EXTENSION}')
```
